```python
import requests
from bs4 import BeautifulSoup

def scrape_website(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Print the title of the webpage
        print(f"Title: {soup.title.text}\n")

        # Print all the paragraph text
        paragraphs = soup.find_all('p')
        print("Paragraphs:")
        for paragraph in paragraphs:
            print(paragraph.text)
```

```python
import requests
from bs4 import BeautifulSoup

def scrape_website(url):
    response = requests.get(url)
    if response.status_code == 200:
        soup = BeautifulSoup(response.text, 'html.parser')

        # Print the title of the webpage
        print(f"Title: {soup.title.text}\n")

        # Print all the paragraph text
        paragraphs = soup.find_all('p')
        print("Paragraphs:")
        for paragraph in paragraphs:
            print(paragraph.text)

        print("\n")

        # Print URLs of all the links on the page
        links = soup.find_all('a', href=True)
        print("Links:")
        for link in links:
            print(link['href'])
```

```python
# Example usage
scrape_website("https://www.coursera.org/browse/data-science")
```

```python
# Install required libraries
!pip install ipywidgets requests beautifulsoup4

# Import necessary libraries
import ipywidgets as widgets
from IPython.display import display, HTML
from bs4 import BeautifulSoup
import requests

# Define the scraping function
def scrape_website(url):
    try:
        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.text, 'html.parser')
            return soup.title.text
        else:
            return f"Error: Unable to fetch the webpage (Status Code: {response.status_code})"
    except Exception as e:
        return f"Error: {str(e)}"

# Create widgets
url_input = widgets.Text(placeholder='Enter URL', description='URL:')
scrape_button = widgets.Button(description='Scrape')
result_output = widgets.Output()

# Define the event handler for the button click
def on_scrape_button_click(b):
    result_output.clear_output()
    with result_output:
        url = url_input.value
        result = scrape_website(url)
        print(result)

# Attach the event handler to the button
scrape_button.on_click(on_scrape_button_click)

# Display the UI
display(url_input, scrape_button, result_output)
```
